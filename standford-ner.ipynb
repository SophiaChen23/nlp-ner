```python
import logging
import csv
import pickle
import numpy as np
doc_count = 0
complete_docs = []
import nltk.tag.stanford  as st
st = st.StanfordNERTagger('/home/sophia/Downloads/stanford-ner-2018-02-27/classifiers/english.all.3class.distsim.crf.ser.gz','/home/sophia/Downloads/stanford-ner-2018-02-27/stanford-ner-3.9.1.jar')
with open('petitions_complete.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        new={}
        doc = row.copy()
        new['title_body']=doc['body']+doc['title']
        new['doc_id']=doc["doc_id"]
        temp=st.tag(new['title_body'].split())
        for item in temp:
            if item[1]!='O' and item[1] not in new:
                new[item[1]]=[item[0]]
            elif item[1]!='O' and item[1]  in new:
                 new[item[1]].append(item[0])
        
        complete_docs.append(new)
for x in complete_docs:
    print(x)
```
